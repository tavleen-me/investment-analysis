### 1. A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type. This chart should make it clear that a certain funding type (FT) is best suited for Teclov.

```python
# import packages to plot a graph
import matplotlib.pyplot as plt
import seaborn as sns

# set seaborn style
sns.set_style("whitegrid")
```

```python
# Limiting our master frame to only 3 types of funding
master_frame = master_frame[master_frame['funding_round_type'].isin(['venture', 'seed', 'private_equity'])]

# setting the size of the figure of the graph
plt.figure(figsize=(11, 5))

# 1st subplot for the number of investments
plt.subplot(1, 2, 1)
sns.countplot(x="funding_round_type", data = master_frame)
plt.title("Fraction of total Investments")

# 2nd subplot for the Average number of investments
plt.subplot(1, 2, 2)
sns.barplot(y='raised_amount_usd', x="funding_round_type", data=master_frame, estimator = np.mean)
plt.title("Average Investment")

# drawing the lines at y = 0, that will span the xrange
plt.axhline(y=5000000, linewidth=3, color = 'c')
plt.axhline(y=15000000, linewidth=3, color = 'c')

# display the plot
plt.show()
```
